```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import tensorflow as tf
def test_image(model, image_path):

  # Load the image
  img = cv2.imread(image_path)
  plt.imshow(img)
  img = cv2.resize(img, (224, 224))
  #plt.imshow(img)
  img = img.astype(np.float32) / 255.
  image = np.stack(img, axis=0)
  print(image.shape)
  image = np.array([image])
  # Predict the class of the image
  prediction = model.predict(image)

  # Get the predicted class label
  predicted_class = np.argmax(prediction)

  # Return the predicted class label
  return predicted_class
```

```python
# Load the model
model1 = tf.keras.models.load_model('C:/Users/nani1/OneDrive/Desktop/Densenet169/60 epochs/densenet169.h5')
```

```python
image_path = 'C:/Users/nani1/OneDrive/Desktop/oral cancer dataset/Oral Cancer/oral_normal/oral_normal_0004.jpg'

# Test the image
predicted_class = test_image(model1, image_path)

if predicted_class==1:
  label='Cancer'
else:
  label='Not Cancer'
print(f"Predicted class label: {label}")
```
